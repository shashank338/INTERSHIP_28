# 1. Scrape the details of most viewed videos on YouTube from Wikipedia.


Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:


A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
# Access chrome browser using webdrive
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(3)


#Access the website
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(3)


# creating empty dictionary
Most_Viewed = {}
Most_Viewed['Rank'] = []
Most_Viewed['Name'] = []
Most_Viewed['Artist'] = []
Most_Viewed['Upload_date'] = []
Most_Viewed['Views'] = []
time.sleep(3)


a = 1 #counter value


# Scraping the all data
for i in range(0,30):
    time.sleep(3)
    
    try:
        #rank
        rank = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[1]".format(a))
        Most_Viewed['Rank'].append(rank.text)
        time.sleep(2)
            
        #name
        name = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[2]".format(a))
        Most_Viewed['Name'].append(name.text)
        time.sleep(2)
        
        #artist
        artist = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[3]".format(a))
        Most_Viewed['Artist'].append(artist.text)
        time.sleep(2)
    
        #upload_date
        uploadDate = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[5]".format(a))
        Most_Viewed['Upload_date'].append(uploadDate.text)
        time.sleep(2)
    
        #views
        view = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[4]".format(a))
        Most_Viewed['Views'].append(view.text)
        time.sleep(2)
    except NoSuchElementException as n:
        print("Exception",n)
   
    # incresing counter variable
    a += 1


Most_Viewed = pd.DataFrame(Most_Viewed)
Most_Viewed
```

# 6. Scrape the details of top 100 songs on billiboard.com.

Url = https:/www.billboard.com/

You have to find the following details:

A) Song name

B) Artist name

C) Last week rank

D) Peak rank

E) Weeks on board

```python
driver = webdriver.Chrome("chromedriver.exe")
url =  "https://www.billboard.com/charts/hot-100"
driver.get(url)

Song_Name =[]
Singer=[]
rank=[]
Last_Week=[]
Weeks_on_board=[]

#scraping the Song_Name 
son=driver.find_elements_by_xpath("//li[@Class='lrv-u-width-100p']")
for i in son:
    if i.text is None :
        Song_Name.append("--") 
        Singer.append("--") 
        rank.append("--") 
        Last_Week.append("--") 
        Weeks_on_board.append("--") 
    else:
        Song_Name.append(i.text.split('\n')[0])
        Singer.append(i.text.split('\n')[1])
        rank.append(i.text.split('\n')[2])
        Last_Week.append(i.text.split('\n')[3])
        Weeks_on_board.append(i.text.split('\n')[4])


BillBOARD_TOP100=pd.DataFrame([])
BillBOARD_TOP100['SONG NAME']=Song_Name
BillBOARD_TOP100['PEAK RANK']=rank
BillBOARD_TOP100['ARTIST NAME']=Singer
BillBOARD_TOP100['LAST WEEK RANK']=Last_Week
BillBOARD_TOP100['WEEKS ON BOARD']=Weeks_on_board

BillBOARD_TOP100
```

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://archive.ics.uci.edu/"
driver.get(url)
time.sleep(6)
driver.maximize_window()

view_dataste= driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]/a")
view_dataste.click()
time.sleep(4)

# creating empty lists for scraping data
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]

#scraping Dataset name
data_name= driver.find_elements_by_xpath("//p[@class='normal']/b/a[1]")
for i in data_name:
    if i.text is None :
        Dataset_name.append("--") 
    else:
        Dataset_name.append(i.text)

time.sleep(4)        
#Scraping data Type
data_type=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[2]/p")
for i in data_type[1:]:
    if i.text is None :
        Data_type.append("--") 
    else:
        Data_type.append(i.text)

time.sleep(4) 
#Scraping Task
task=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[3]/p")
for i in task[1:]:
    if i.text is None :
        Task.append("--") 
    else:
        Task.append(i.text)
        
time.sleep(4) 
#Scraping Attribute_type
attribute_type=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[4]/p")
for i in attribute_type[1:]:
    if i.text is None :
        Attribute_type.append("--") 
    else:
        Attribute_type.append(i.text)

time.sleep(4)         
#Scraping No_of_instances
no_of_instances=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[5]/p")
for i in no_of_instances[1:]:
    if i.text is None :
        No_of_instances.append("--") 
    else:
        No_of_instances.append(i.text)

time.sleep(4)         
#Scraping No_of_attribute
no_of_attribute=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[6]/p")
for i in no_of_attribute[1:]:
    if i.text is None :
        No_of_attribute.append("--") 
    else:
        No_of_attribute.append(i.text)

time.sleep(4) 
#Scraping Year
year=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[7]/p")
for i in year[1:]:
    if i.text is None :
        Year.append("--") 
    else:
        Year.append(i.text)
        
Datasets=pd.DataFrame({})
Datasets['DATASET NAMES']= Dataset_name
Datasets['DATA TYPE']= Data_type
Datasets['TASK']= Task
Datasets['ATTRIBUTE TYPE']= Attribute_type
Datasets['NO OF INSTANCES']= No_of_instances
Datasets['NO OF ATTRIBUTE']= No_of_attribute
Datasets['YEAR']= Year
Datasets
```
